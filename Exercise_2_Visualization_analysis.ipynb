```python
import matplotlib.pyplot as plt

# Dataset
fruits = ["Apples", "Bananas", "Oranges", "Grapes"]
sales = [50, 80, 40, 70]

# Create bar chart
plt.figure(figsize=(8,5))
plt.bar(fruits, sales)

# Labels and title
plt.title("Fruit Sales")
plt.xlabel("Fruit Type")
plt.ylabel("Number of Sales")

# Display chart
plt.show()
```

# 1. The graph allows business managers to compare quickly fruit sales between categories

# 2. Visualization dimension using the Visualization Wheel. The visualization is undimensional because it mainly represents one quantitative variable: sales.

# 3. Drawbacks Following Cairo and Tufte Principles

One drawback of this visualization is that it only shows total sales and does not explain why some fruits sell more than others.

Another limitation is that using too many colors, shadows, or 3D effects could reduce clarity and create chart junk.

According to Edward Tufte’s principles, visualizations should maximize clarity and minimize unnecessary graphical elements.

# 4. Graphical Variables Used

The graph uses position and size (bar height) to represent numerical values.

These graphical variables are appropriate because viewers can easily compare differences between categories.

Color is used only minimally to avoid distraction and improve readability.

# Conclusion

This visualization is effective because it presents information clearly and simply. The chart follows several good visualization practices discussed in class, including readability, simplicity, and avoidance of misleading graphical elements.
